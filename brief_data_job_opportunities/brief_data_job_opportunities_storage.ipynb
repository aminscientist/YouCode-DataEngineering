**Connection à SQL server**

```python
import pyodbc

driver = 'SQL Server'
server = 'LAPTOP-EEA5QPRQ\SQLEXPRESS'
database = 'EmploymentInsightsDB'

conn = pyodbc.connect(f'''
                        DRIVER={driver};
                        SERVER={server};
                        DATABASE={database};
                        Trusted_Connection=yes
                       ''')

cursor = conn.cursor()
```

**Création des tables**

```python
# Créer la table Types
query_types = '''
CREATE TABLE Types (
    id_type INTEGER PRIMARY KEY,
    job_type VARCHAR(255)
);
'''
conn.execute(query_types)

# Créer la table Experiences
query_experiences = '''
CREATE TABLE Experiences (
    id_experience INTEGER PRIMARY KEY,
    experience_level VARCHAR(255)
);
'''
conn.execute(query_experiences)

# Créer la table Currencies
query_currencies = '''
CREATE TABLE Currencies (
    id_currency INTEGER PRIMARY KEY,
    currency VARCHAR(255)
);
'''
conn.execute(query_currencies)

# Créer la table Modes
query_modes = '''
CREATE TABLE Modes (
    id_mode INTEGER PRIMARY KEY,
    work_mode VARCHAR(255)
);
'''
conn.execute(query_modes)

# Créer la table Companies
query_companies = '''
CREATE TABLE Companies (
    id_company INTEGER PRIMARY KEY,
    company VARCHAR(255)
);
'''
conn.execute(query_companies)

# Créer la table Countries
query_countries = '''
CREATE TABLE Countries (
    id_country INTEGER PRIMARY KEY,
    country VARCHAR(255)
);
'''
conn.execute(query_countries)

# Créer la table Job_offers
query_job_offers = '''
CREATE TABLE Job_offers (
    id_job INTEGER PRIMARY KEY,
    job_title VARCHAR(255),
    location VARCHAR(255),
    salary VARCHAR(50),
    salary_conditions BIT,
    id_company INTEGER,
    id_country INTEGER,
    id_type INTEGER,
    id_experience INTEGER,
    id_currency INTEGER,
    id_mode INTEGER
);
'''
conn.execute(query_job_offers)

# Créer la table Facilities
query_facilities = '''
CREATE TABLE Facilities (
    id_facility INTEGER PRIMARY KEY,
    facility VARCHAR(255)
);
'''
conn.execute(query_facilities)

# Créer la table Requirements
query_requirements = '''
CREATE TABLE Requirements (
    id_requirement INTEGER PRIMARY KEY,
    requirement VARCHAR(255)
);
'''
conn.execute(query_requirements)
```

**Création des tables pontes Job_facilities et Job_requirements**

```python
# Créer la table ponte Job_facilities
requete_creation_table_job_facilities = """
CREATE TABLE Job_facilities (
    id_job INTEGER,
    id_facility INTEGER,
    PRIMARY KEY (id_job, id_facility),
    FOREIGN KEY (id_job) REFERENCES Job_offers (id_job),
    FOREIGN KEY (id_facility) REFERENCES Facilities (id_facility)
)
"""

# Exécuter la requête pour créer la table Job_facilities
cursor.execute(requete_creation_table_job_facilities)

# Créer la table ponte Job_requirements
requete_creation_table_job_requirements = """
CREATE TABLE Job_requirements (
    id_job INTEGER,
    id_requirement INTEGER,
    PRIMARY KEY (id_job, id_requirement),
    FOREIGN KEY (id_job) REFERENCES Job_offers (id_job),
    FOREIGN KEY (id_requirement) REFERENCES Requirements (id_requirement)
)
"""

# Exécuter la requête pour créer la table Job_requirements
cursor.execute(requete_creation_table_job_requirements)
```
