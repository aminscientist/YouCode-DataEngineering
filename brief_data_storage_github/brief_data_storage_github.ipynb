**CrÃ©ation d'une nouvelle dataframe pour stocker la liste unique des topics avec leurs index**

```python
df['topics'].replace('[no topics]', "['no topics']", inplace = True)
df
```

```python
df['topics'] = df['topics'].apply(eval)
expanded_data = []
for index, row in df.iterrows():
    id_repo = row['id_repo']
    topics = row['topics']
    
    for topic in topics:
        expanded_data.append({'id_repo': id_repo, 'topic': topic})
    if len(topics) == 0:
        expanded_data.append({'id_repo': id_repo, 'topic': None})

df_topics_per_repos = pd.DataFrame(expanded_data)
```

```python
df_topics_per_repos
```

```python
unique_topics = df_topics_per_repos['topic'].drop_duplicates().reset_index(drop=True)
df_topics = pd.DataFrame({'topic': unique_topics})
df_topics['id_topic'] = df_topics.index + 1
df_topics
```

```python
df_topics_repo = df_topics_per_repos.merge(df_topics, on='topic', how='left')
df_topics_repo
```

```python
import pyodbc

driver = 'SQL Server'
server = 'LAPTOP-EEA5QPRQ\SQLEXPRESS'
database = 'GitHubDB'

conn = pyodbc.connect(f'''
                        DRIVER={driver};
                        SERVER={server};
                        DATABASE={database};
                        Trusted_Connection=yes
                       ''')

cursor = conn.cursor()
```

**CrÃ©ation des tables**

```python
# CrÃ©ez la table Languages
cursor.execute('''
    CREATE TABLE Languages (
        id_language INT PRIMARY KEY,
        language VARCHAR(255)
    )
''')

# CrÃ©ez la table Repositories
cursor.execute('''
    CREATE TABLE Repositories (
        id_repo INT PRIMARY KEY,
        url VARCHAR(255),
        full_name VARCHAR(255),
        description TEXT,
        forks INT,
        watchers INT,
        issues INT,
        stars INT,
        date_created DATE,
        last_push DATE,
        description_translated TEXT,
        id_language INT,
        FOREIGN KEY (id_language) REFERENCES Languages(id_language)
    )
''')


# CrÃ©ez la table Contributors
cursor.execute('''
    CREATE TABLE Contributors (
        id_contributor INT PRIMARY KEY,
        contributor VARCHAR(255)
    )
''')

# CrÃ©ez la table Topics
cursor.execute('''
    CREATE TABLE Topics (
        id_topic INT PRIMARY KEY,
        topic VARCHAR(255)
    )
''')

# CrÃ©ez la table Contributions
cursor.execute('''
    CREATE TABLE Contributions (
        id_repo INT,
        id_contributor INT,
        contributions INT,
        FOREIGN KEY (id_repo) REFERENCES Repositories(id_repo),
        FOREIGN KEY (id_contributor) REFERENCES Contributors(id_contributor),
        PRIMARY KEY (id_repo, id_contributor)
    )
''')

# CrÃ©ez la table Topics_repo
cursor.execute('''
    CREATE TABLE Topics_repo (
        id_topic INT,
        id_repo INT,
        FOREIGN KEY (id_topic) REFERENCES Topics(id_topic),
        FOREIGN KEY (id_repo) REFERENCES Repositories(id_repo),
        PRIMARY KEY (id_topic, id_repo)
    )
''')
```
